**Polinomio dividido por una de sus raices**

```
from sympy import Symbol
from sympy import div

x = Symbol('x')

p = (x**10) - (81* x**9) + (2734*x*8) - (49266*x*7) + (495397*x*6) - (2437029*x*5) - (16164*x**4) + (65582136*x**3) - (331831728*x**2) + (684618480*x) - 514382400

p, r = div(p,  x-3)

print(p)
print(r)

p, r = div(p,  x-6)

print(p)
print(r)

p, r = div(p,  x-9)

print(p)
print(r)

p, r = div(p,  x-12)

print(p)
print(r)

p, r = div(p,  x-15)

print(p)
print(r)

p, r = div(p,  x-18)

print(p)
print(r)

p, r = div(p,  x-7)

print(p)
print(r)

p, r = div(p,  x+5)

print(p)
print(r)


p, r = div(p,  x-14)

print(p)
print(r)


p, r = div(p,  x-2)

print(p)
print(r)
```

## SymPy’s Symbolic and Numercical Complex Evaluations
- https://docs.sympy.org/latest/modules/evalf.html](https://)
- x = x1 + I*x2,y = y1 + I*y2, z = z1 + I*z2, x*y*z

** Example **

$ x = x1 + lx2+lx3,y = y1 + ly2+ly3, z = z1 + lz2+lz3, x*y+z $


image.png

```
# Ejemplo en Sympy
from sympy import *
x1, x2, x3, y1, y2, y3, z1, z2, z3 = symbols("x1 x2 x3 y1 y2 y3 z1 z2 z3", real=True)  
x = x1 + I*x2 + I*x3
y = y1 + I*y2 + I*y3
z = z1 + I*z2 + I*z3

print(x*y+z)
print(expand(x*y+z))
print(expand((x*y)+z))
print(expand(x*(y+z)))

w=N(1/(pi + I), 20)
print('w=',w)
```
